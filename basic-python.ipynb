```python
import time
import sys
h=3
m=53
s=0
print("\033[1;31m")
while True:
    try:
        sys.stdout.write("\r{:02d}:{:02d}:{:02d}".format(h,m,s))
        sys.stdout.flush()
        time.sleep(1)
        s+=1
        if s==60:
            s=0
            m+=1
        if m==60:
            m=0
            h+=1
        if h==13:
            h=1
            m=0
            s=0
    except KeyboardInterrupt:
        print('\nStopped.')
        break
    
```

```python
names=['raM','RAJ','rose','JaCk','PaVi']
marks=[[30,40,50],[66,77,55],[33,78,90],[87,97,57],[47,88,44]]
''' 
less than 40-F
60-80 A
more than 80 S
41-60 B
'''
pos=0
for i in range(5):
    avg=sum(marks[i])//3
    if avg>80:
        grades='S'
    elif avg>=60:
        grades='A'
    elif avg>40:
        grades='B'
    else:
        grades='F'
    print("{}. {} has scored {}% - {} grade".format(pos+1,names[i].title(),avg,grades))
    pos+=1
```

```python
#scheduling
import random
a=int(input('Enter no. of teams: '))
teams=[]
for i in range(a):
    t=input('Team name: ')
    teams.append(t)
m=int(input('No. of meet b/w teams: '))

matches=[]
for i in range(0,a-1):
    for j in range(i+1,a):
        for k in range(m):
            matches.append([teams[i],teams[j]])
            
random.shuffle(matches)
pos=1
for i in matches:
    print('MATCH {} : {} vs {}'.format(pos,i[0],i[1]))
    pos+=1
```

```python
#program for FLAMES
a1=input("Enter boy name: ")
b1=input("Enter girl name: ")
a=list(a1)
b=list(b1)

for i in range(len(a)):
    for j in range(len(b)):
        if a1[i]==b[j]:
            a[i]='2'
            b[j]='2'
cnt=0
for i in a:
    if i!='2':
        cnt+=1
for i in b:
    if i!='2':
        cnt+=1
        
idx=0
res=list('FLAMES')
for i in range(5):
    idx=(idx+(cnt-1))%len(res)
    res.pop(idx)
print(res[0])
```
