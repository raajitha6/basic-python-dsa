```python
for i in range(100,1000):
    prime=True
    for j in range(2,i):
        if(i%j==0):
            prime=False
            break
    if(prime):
        print(i, end=" ")
```

```python
legs=int(input("no. of legs:"))
heads=int(input("no. of heads:"))
flag=False
for cows in range(heads+1):
    hens=heads-cows
    cal_legs=cows*4+hens*2
    if cal_legs==legs:
        flag=True
        break
if flag:
    print("Cows:",cows)
    print("Hens:",hens)
else:
    print("No solution")
```

```python
a=input("Enter password:")
uc=0
lc=0
dg=0
sp=0
if len(a)>7:
    for i in a:
        if i.isupper():
            uc+=1
        elif i.islower():
            lc+=1
        elif i.isdigit():
            dg+=1
        else:
            sp+=1
    if uc>0 and lc>0 and dg>0 and sp>0:
        print("STRONG")
    else:
        print("WEAK")
else:
    print("WEAK due to less no. of chars")
```

```python
import time
import sys
h=3
m=53
s=0
print("\033[1;31m")
while True:
    try:
        sys.stdout.write("\r{:02d}:{:02d}:{:02d}".format(h,m,s))
        sys.stdout.flush()
        time.sleep(1)
        s+=1
        if s==60:
            s=0
            m+=1
        if m==60:
            m=0
            h+=1
        if h==13:
            h=1
            m=0
            s=0
    except KeyboardInterrupt:
        print('\nStopped.')
        break
    
```
