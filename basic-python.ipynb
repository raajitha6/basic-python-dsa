```python
names=['raM','RAJ','rose','JaCk','PaVi']
marks=[[30,40,50],[66,77,55],[33,78,90],[87,97,57],[47,88,44]]
''' 
less than 40-F
60-80 A
more than 80 S
41-60 B
'''
pos=0
for i in range(5):
    avg=sum(marks[i])//3
    if avg>80:
        grades='S'
    elif avg>=60:
        grades='A'
    elif avg>40:
        grades='B'
    else:
        grades='F'
    print("{}. {} has scored {}% - {} grade".format(pos+1,names[i].title(),avg,grades))
    pos+=1
```

```python
#scheduling
import random
a=int(input('Enter no. of teams: '))
teams=[]
for i in range(a):
    t=input('Team name: ')
    teams.append(t)
m=int(input('No. of meet b/w teams: '))

matches=[]
for i in range(0,a-1):
    for j in range(i+1,a):
        for k in range(m):
            matches.append([teams[i],teams[j]])
            
random.shuffle(matches)
pos=1
for i in matches:
    print('MATCH {} : {} vs {}'.format(pos,i[0],i[1]))
    pos+=1
```

#### LINKED LIST :-

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

class LinkedList:
    def __init__(self):
        self.head=None

    def insert(self,data):
        newnode=Node(data)
        if self.head==None:
            self.head=newnode
        else:
            temp=self.head
            while temp.next:
                temp=temp.next
            temp.next=newnode

    def display(self):
        temp=self.head
        while temp:
            print(temp.data,"->",end=" ")
            temp=temp.next
    

l=LinkedList()
l.insert(100)
l.insert(200)
l.insert(300)
l.insert(400)
l.insert(500)
l.display()
```
